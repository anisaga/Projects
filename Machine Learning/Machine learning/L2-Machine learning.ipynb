```python
cols=['Time','Amount','V2','V3','V5','V11','V12','V14','V17','V26']
X=credit_card[cols]
Y=credit_card[['Class']] #feature engineering
```

```python
X.head() #first few columns of independent features
```

```python
Y.head() #first few columns of dependent features
```

```python
print('Independent Feature Set Shape : ', X.shape)
print('Dependent Feature Shape       : ', Y.shape) #shape of column
```

```python
from imblearn.under_sampling import NearMiss
nm_sampler=NearMiss()
x_sampled,y_sampled=nm_sampler.fit_resample(X,Y)

print('Data   : ',x_sampled.shape)
print('Labels : ',y_sampled.shape) #under_sample because our dependent features not balanced 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.25, random_state = 5) #split the columns
```

```python
print('Training Data Shape   : ', x_train.shape)
print('Training Labels Shape : ', y_train.shape)
print('Testing Data Shape    : ', x_test.shape)
print('Testing Labels Shape  : ', y_test.shape)#data shape
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', max_depth = 10)
classifier.fit(x_train, y_train)

pred_train = classifier.predict(x_train)
pred_test  = classifier.predict(x_test)#1st model
```

```python
from sklearn.metrics import accuracy_score
print('Training set accuracy : ',accuracy_score(y_train,pred_train))
print('Testing set accuracy : ',accuracy_score(y_test,pred_test)) #model score
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 30, criterion = 'entropy', max_depth = 10)
classifier.fit(x_train, y_train)#2nd model
```

```python
pred_train=classifier.predict(x_train)
pred_test=classifier.predict(x_test)

from sklearn.metrics import accuracy_score
print('Training set accuracy : ',accuracy_score(y_train,pred_train))
print('Testing set accuracy : ',accuracy_score(y_test,pred_test))#pred and score of model
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=3)
classifier.fit(x_train,y_train)#3 model
```

```python
pred_train=classifier.predict(x_train)
pred_test=classifier.predict(x_test)

from sklearn.metrics import accuracy_score
print('Training set accuracy : ',accuracy_score(y_train,pred_train))
print('Testing set accuracy : ',accuracy_score(y_test,pred_test))#score and prediction
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()#hyper tuning
```

```python
search_params = {
    'n_estimators' : [10,20,30,40,50],
    'max_depth'    : [3,4,5,6,8,10],
    'max_features' : [3,4,5,6]
}#parameters
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator = classifier,
                           param_grid = search_params,
                           scoring = ['r2'],
                           cv = 5,
                           refit = 'r2',
                           verbose = 1)
grid_search.fit(x_train, y_train) #hyperparameters
```
