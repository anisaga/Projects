```python
#segmentation based on popularity with danceability and liveness
X=df[['popularity','danceability','liveness']].values
X[:5] #extract features 
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3)
km.fit(X) #assigning cluster
```

```python
y_kmeans = km.predict(X)
y_kmeans #predict
```

```python
km.cluster_centers_ #cluster centers
```

```python
km.inertia_ #wcss
```

```python
# Extracting all data points belonging to cluster 0
X[y_kmeans == 0]
```

```python
#Extracting data points belonging to cluster 0
X[y_kmeans == 0,0]
```

```python
wcss = []
for k in range(1,21):
    km = KMeans(n_clusters = k)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
k = range(1,21)
plt.figure(figsize = (16,8), dpi = 200)
plt.plot(k, wcss)

plt.xticks(k)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #extract best value of clusters
```

```python
# Best model
km_best = KMeans(n_clusters = 4)
km_best.fit(X)
```

```python
y_kmeans = km_best.predict(X)
y_kmeans #performance
```

```python
centroids = km_best.cluster_centers_
centroids #centroids
```
