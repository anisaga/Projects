```python
data=data.drop(['date','id','zipcode'],axis=1) #drop this column which is least significant
```

```python
data.columns #columns in data
```

```python
cols=['bedrooms','bathrooms','sqft_living','grade','lat','long','view','condition']
X=data[cols]
Y=data[['price']] #feature engineering
```

```python
X.head(10) #first feew columns of independent features
```

```python
Y.head() #first few columns of dependent features
```

```python
print('Independent Feature Set Shape : ', X.shape)
print('Dependent Feature Shape       : ', Y.shape)#check shape of our feattures
```

```python
# Splitting The Data into training and test sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = 2)
```

```python
print('Training Data Shape   : ', x_train.shape)
print('Training Labels Shape : ', y_train.shape)
print('Testing Data Shape    : ', x_test.shape)
print('Testing Labels Shape  : ', y_test.shape) #check shape of our data after train and test splitting
```

```python
# Scaling
cols = ['sqft_living','lat','long','grade']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Fit on training set
scaler.fit(x_train[cols])

# Tranform training and testing sets
x_train[cols] = scaler.transform(x_train[cols])
x_test[cols]  = scaler.transform(x_test[cols])
```

```python
x_train.head(10) #after scaling check data
```

```python
x_test.head(10) #check testing data after scaling
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)#1st reg model
```

```python
regressor.intercept_          #B0 parameter
```

```python
regressor.coef_   #b1x parameter
```

```python
# Evaluation
print('Training Set Score : ', regressor.score(x_train, y_train))
print('Testing Set Score  : ', regressor.score(x_test, y_test))
```

```python
from sklearn.metrics import r2_score

pred_train = regressor.predict(x_train)
pred_test  = regressor.predict(x_test)

print('Training Set Score : ', r2_score(y_train, pred_train))
print('Testing Set Score  : ', r2_score(y_test, pred_test)) #check model performence

```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion = 'squared_error', max_depth = 13)
regressor.fit(x_train, y_train) #2 reg model
```

```python
pred_train = regressor.predict(x_train)
pred_test  = regressor.predict(x_test) #prediction 
```

```python
from sklearn.metrics import r2_score
print('Training Set Score : ', r2_score(y_train, pred_train))
print('Testing Set Score  : ', r2_score(y_test, pred_test)) #model performence
```

```python
from sklearn.neighbors import KNeighborsRegressor
regressor = KNeighborsRegressor(n_neighbors =7)
regressor.fit(x_train, y_train) #3 reg model
```

```python
pred_train = regressor.predict(x_train)
pred_test  = regressor.predict(x_test) #prediction

from sklearn.metrics import r2_score
print('Training Set Score : ', r2_score(y_train, pred_train))
print('Testing Set Score  : ', r2_score(y_test, pred_test)) #model performence
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 40, max_depth = 6, max_features =5)
regressor.fit(x_train , y_train) #3 reg model
```
