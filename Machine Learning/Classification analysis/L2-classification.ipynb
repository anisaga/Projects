```python
cols=['Area','Perimeter','MajorAxisLength','MinorAxisLength','EquivDiameter','ConvexArea']
X=data[cols]
Y=data[['Class']]#feature engineering
```

```python
X.head()#x columns
```

```python
Y.head()#y columns
```

```python
# Min Max Scaler
cols=[ 'Area','Perimeter','MajorAxisLength','MinorAxisLength','ConvexArea','EquivDiameter']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X[cols] = scaler.fit_transform(X[cols])
X.head(10)
```

```python
#Splitting The Data into training and test sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = 2)
```

```python
print('Training Data Shape   : ', x_train.shape)
print('Training Labels Shape : ', y_train.shape)
print('Testing Data Shape    : ', x_test.shape)
print('Testing Labels Shape  : ', y_test.shape)#data shape
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', max_depth = 10)
classifier.fit(x_train, y_train)

pred_train = classifier.predict(x_train)
pred_test  = classifier.predict(x_test)#1st model
```

```python
from sklearn.metrics import accuracy_score
print('Training set accuracy : ',accuracy_score(y_train,pred_train))
print('Testing set accuracy : ',accuracy_score(y_test,pred_test)) #model score
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 30, criterion = 'entropy', max_depth = 10)
classifier.fit(x_train, y_train)#2nd model
```

```python
pred_train=classifier.predict(x_train)
pred_test=classifier.predict(x_test)

from sklearn.metrics import accuracy_score
print('Training set accuracy : ',accuracy_score(y_train,pred_train))
print('Testing set accuracy : ',accuracy_score(y_test,pred_test))#pred and score of model
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=3)
classifier.fit(x_train,y_train)#3 model
```

```python
pred_train=classifier.predict(x_train)
pred_test=classifier.predict(x_test)

from sklearn.metrics import accuracy_score
print('Training set accuracy : ',accuracy_score(y_train,pred_train))
print('Testing set accuracy : ',accuracy_score(y_test,pred_test))#score and prediction
```

```python
from lightgbm import LGBMClassifier
classifier = LGBMClassifier(n_estimators = 50)
classifier.fit(x_train, y_train)#4th model
```

```python
pred_train = classifier.predict(x_train)
pred_test = classifier.predict(x_test)

print('Training Set Accuracy : ', accuracy_score(y_train, pred_train))
print('Testing Set Accuracy  : ', accuracy_score(y_test, pred_test))#prediction and score
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()#hyper tuning
```

```python
search_params = {
    'n_estimators' : [10,20,30,40,50],
    'max_depth'    : [3,4,5,6,8,10],
    'max_features' : [3,4,5,6]
}#parameters
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator = classifier,
                           param_grid = search_params,
                           scoring = ['r2'],
                           cv = 5,
                           refit = 'r2',
                           verbose = 1)
grid_search.fit(x_train, y_train)#tune the parameters
```

```python
best_model = grid_search.best_estimator_
best_model#best model
```
