#****1.Do universities with a high number of applications are the preferred ones by students? ***
     2.Do students prefer universities that have a high rate of admission? in other words, do students prefer a university where it is easier for them to be admitted?

```python
plt.figure(figsize=(12,5),dpi=200) #correlation between applicant and admission along wioth enrolment
sns.scatterplot(x='Applicants total',y='Admissions total',hue='Enrolled total',data=df1)
```

##Universities with more applicants are most preferred one and make more enrollments
###Also universities with more admissions have high admittence rate are preferred by students and have more application

####3.	Do students prefer public or private universities?

```python
plt.figure(figsize=(12,5),dpi=200) #find no of public and private university
sns.countplot(x='Control of institution',data=df1) #more private university available
```

```python
plt.figure(figsize=(12,5),dpi=200) #total enrollment on basis of institution
sns.barplot(x='Control of institution',y='Enrolled total',data=df1)
```

#more students enrolled in public universities

```python
plt.figure(figsize=(12,5),dpi=200)#total applicants received on institute types
sns.barplot(x='Control of institution',y='Applicants total',data=df1)
```

#More students applied for public universities

##4.Do students prefer universities with low tuition and fees?

```python
plt.figure(figsize=(12,5),dpi=200) #application received on basis of tuition fee
sns.scatterplot(x='Applicants total',y='Tuition and fees, 2013-14',hue='Enrolled total',data=df1)
```

##when tuition fees are high ,applicants are less and enrollment is also less

```python
plt.figure(figsize=(12,5),dpi=200) #fees received by institute
sns.barplot(x='Control of institution',y='Tuition and fees, 2013-14',data=df1)
```

##Tuition fees higher in private so most preferr public universities 

##5.	Do students prefer a university for its low on-campus cost of living?

```python
plt.figure(figsize=(12,5),dpi=200) #based opn campus living cost and applicant
sns.scatterplot(x='Total price for in-state students living on campus 2013-14',y='Applicants total',hue='Enrolled total',data=df1)
```

#students not preferred high on campus cost of living and more enrolment rate in lower tuition fee

6.	Do students prefer universities from highly populated states?

```python
populated_states=['California','Texas','Newyork','Ohio','Florida']
df1['populated']=df1['state'].apply(lambda x: 'High' if x in populated_states else 'Low' )
df1.head() #find high and low populated state
```

```python
plt.figure(figsize=(12,5),dpi=200) #find applicants and population
sns.barplot(x='populated',y='Applicants total',data=df1)
```

#students preferred higly populated states than low populated states

#7.	Do students prefer a university because it belongs to a state with a high GDP per capita?

```python
gdp=df1.sort_values(by='gdp_million',ascending=False)
gdp.head(20)#find gdp is high or low
```

```python
high_gdp=['Massachusetts','Connecticut','New york','California','New jersey']

df1['gdp']=df1['state'].apply(lambda x:'High gdp' if x in high_gdp else 'Low gdp') #create columns with label of high or low gdp
```
