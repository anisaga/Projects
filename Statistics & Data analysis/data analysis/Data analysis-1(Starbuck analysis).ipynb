```python
starbucks1=pd.read_csv('starbucks-menu-nutrition-drinks.csv')
starbucks1.head(10) #Starbucks nutrition drinks file
```

```python
starbucks1.info() #info of starbucks nutrition drinks 
```

```python
starbucks1.nunique() #no of unique columns in starbucks1
```

```python
starbucks1.isnull().sum() #total sum of all null values
```

```python
starbucks1.replace({"-":None},inplace=True) #replace - with nan
```

```python
starbucks1.head() #check missing values 
```

```python
starbucks1.isnull().sum() #total null values
```

```python
starbucks1.dropna(inplace=True) #drop null values
```

```python
starbucks1.info() #again check missing values
```

```python
starbucks1=starbucks1.rename(columns={'Fat (g)':'Fat','Carb. (g)':'Carb','Fiber (g)':'Fiber'})
starbucks1.columns #rename columns
```

```python
col=['Calories','Fat','Carb','Fiber','Protein','Sodium']
starbucks1[col]=starbucks1[col].apply(pd.to_numeric)
starbucks1.dtypes #change object to numeric data type
```

```python
plt.figure(figsize=(12,5),dpi=200)
sns.countplot(x='Fat',data=starbucks1) #In starbucks nutrition drink fat is used very minimal 
```

```python
plt.figure(figsize=(12,5),dpi=200)
sns.countplot(x='Protein',data=starbucks1) #protein used in small amount in starbucks nutrition drinks
```

```python
plt.figure(figsize=(12,5),dpi=200)
sns.scatterplot(x='Calories',y='Fat',hue='Carb',data=starbucks1) # calories increase,with high carbs and fat is low
```

```python
plt.figure(figsize=(12,5),dpi=200)
sns.scatterplot(x='Calories',y='Fat',hue='Protein',data=starbucks1) #protein increase with high calories and fat is less
```

```python
cols=['Calories','Fat','Carb','Fiber','Protein','Sodium']
starbucks1[cols].corr()         #calories have higher correlation  
```
