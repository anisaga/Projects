```python
titles=[]
prices=[]
ratings=[]
#scraping through webpage

for x in range(1,21):
    start_time=time.time()
    URL='https://www.flipkart.com/search?q=kids+cycle&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_3_4_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_3_4_na_na_na&as-pos=3&as-type=RECENT&suggestionId=kids+cycle&requestId=2c6968d2-4daa-4aeb-b182-37da6bbee576&as-searchtext=kids'
    page=requests.get(URL)
    soup=BeautifulSoup(page.content,'html.parser')
    
    product_details=soup.find_all('div',class_='_4ddWXP')
    
    for product in product_details:
        #name
        name=product.find('a',class_='s1Q9rs')
        if name is None:
            titles.append(np.nan)
        else:
            titles.append(name.text) 

        price=product.find('div',class_='_30jeq3')
        #price
        if price is None:
            prices.append(np.nan)
        else:
            prices.append(price.text)

        rating=product.find('div',class_='_3LWZlK')
        #rating
        if rating is None:
            ratings.append(np.nan)
        else:
            ratings.append(rating.text)

    end_time=time.time()
    print('page {} completed in {} seconds'.format(x,end_time-start_time))
```

```python
print(len(titles))
print(len(prices))
print(len(ratings))#length of all titles
```

```python
data={'Product_name' : titles,
      'Price'        : prices,
      'Rating'       : ratings}
df=pd.DataFrame(data)
df.head(10) #creating datdaset
```

```python
df.info() #info abooout dataframe
```

```python
df.Rating=df.Rating.astype(float)
df.Price=df.Price.apply(lambda s : float(s.replace('₹','').replace(',','')))

df.info() #convert data type to float
```

```python
df.to_csv('kids_cycle.csv') #data saved as csv
```
